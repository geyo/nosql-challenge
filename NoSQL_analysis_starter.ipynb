### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
## As ratings are stored as strings to account for a 'pass' rating, I used
## "$in" as opposed to "$gte" to search for all values at or above 4.
## All values at or above 4 in this range are "4" and "5". 
query = {
    "LocalAuthorityName": {'$regex':"London"},
    "RatingValue": {'$in': ["4","5"]}
}

## Choose most relevant fields for easier viewing
fields = {'LocalAuthorityBusinessID':1,
          'BusinessName':1,
          'BusinessType':1,
          'RatingValue': 1, 
          'LocalAuthorityName': 1,   
}

## Store query
london_result = establishments.find(query, fields)

# Use count_documents to display the number of documents in the result
count = establishments.count_documents(query)
print(f"There are {count} documents in the result.")

# Display the first document in the results using pprint
pprint(establishments.find_one(query, fields))
```

```python
# Convert the result to a Pandas DataFrame
london_data = [doc for doc in london_result]
london_df = pd.DataFrame(london_data)

# Display the number of rows in the DataFrame
print(f"There are {count} establishments in London that have a rating of 4 or greater.")

# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
## Look up Penang Flavour's lat/long
Penang = establishments.find_one(
    {"BusinessName":{"$regex":"Penang"}}, 
    {"BusinessName":1,
     "geocode.latitude":1,
     "geocode.longitude":1, 
     'LocalAuthorityName': 1})
Penang
```

```python
## Test 'get()' method to access dictionary values (to store lat/long as variable)
Penang.get('geocode').get('longitude')
```

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = Penang.get('geocode').get('latitude') #51.490142
longitude = Penang.get('geocode').get('longitude') #0.08384

## Match Query
query = {'RatingValue':'5', #RatingValue equals 5
                   'geocode.latitude': #Lat within .01 degrees
                         {'$gte': (latitude - degree_search), 
                          '$lt': (latitude + degree_search)
                         },
                    'geocode.longitude': #Long within .01 degrees
                         {'$gte': (longitude - degree_search), 
                          '$lt': (longitude + degree_search)
                         },
                    'scores.Hygiene': #Exclude results w/ no hygiene score. 
                         {'$ne': None}
        }

## Choose fields (for easier viewing)
fields = {'LocalAuthorityBusinessID':1,
          'BusinessName':1,
          'BusinessType':1,
          'RatingValue': 1, 
          'geocode.latitude': 1,
          'geocode.longitude': 1,
          'scores.Hygiene': 1,
          'LocalAuthorityName': 1   
}

## Sort query in ascending order
sort =  [('scores.Hygiene', 1)]

## Limit results to five
limit = 5

## Combine queries
results = list(establishments.find(query, fields).sort(sort).limit(limit))

# Print the results
pprint(results)
```

```python
# Convert result to Pandas DataFrame
## Apply json_normalize to view nested dictionaries as seperate columns. 
normalized_results = pd.json_normalize(results)
near_Penang = pd.DataFrame(normalized_results)
near_Penang.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

## Match Query, with hygiene as zero. 
match_query = {'$match': {'scores.Hygiene': 0}}

## Group Query, by local authority
group_query = {
    '$group': {
        '_id': "$LocalAuthorityName",
        'count_per_local': {'$sum':1}
    }
}

## Sort Query, highest to lowest
sort_query = {
    '$sort': {
        'count_per_local': -1
    }
}

## put pipeline together
pipeline = [match_query, group_query, sort_query]
results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"There are {len(results)} local authorities that have establishments with hygiene scores of zero.")

# Print the first 10 results
pprint(results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygienic_authorities = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"There are {len(hygienic_authorities)} rows in this DataFrame.")

# Display the first 10 rows of the DataFrame
hygienic_authorities.head(10)
```
